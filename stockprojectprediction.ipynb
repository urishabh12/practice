```python
import pandas as pd
import pandas_datareader as pdr
import datetime
from sklearn import linear_model
from joblib import dump, load
```

```python
def get(tickers, startdate, enddate):
  def data(ticker):
    return (pdr.get_data_yahoo(ticker, start=startdate, end=enddate))
  datas = map (data, tickers)
  return(pdr.concat(datas, keys=tickers, names=['Ticker', 'Date']))

tickers = ['AAPL', 'MSFT', 'IBM', 'GOOG']
all_data = pdr.get_data_yahoo(tickers, datetime.datetime(2006, 10, 1), datetime.datetime(2012, 1, 1))
print(all_data)
```

```python
all_data=all_data[all_data.columns[0:4]]
```

```python
temp = all_data[all_data.columns[0:0]]
```

```python
month=[]
day=[]
for i in range(1323):
    a=temp.iloc[i]
    a = str(a)
    a=a[22:27]
    month.append(int(a[:2]))
    day.append(int(a[3:]))
```
